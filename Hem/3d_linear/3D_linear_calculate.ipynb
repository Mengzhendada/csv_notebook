```python
import numpy as np
#import sys
#sys.path.append('/Users/shuojia/CSV/lhapdf/build/lib/python3.8/site-packages')
import lhapdf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
p_cteq6 = lhapdf.mkPDF("cteq6l1", 0)
p_cteq18 = lhapdf.mkPDF("CT18qed",0)
PDF = p_cteq18
```

```python
Hem_file = pd.read_csv('mult_ratio_to_solve_for_F1F2F3.txt',delim_whitespace=True)
Hem_file.head()
```

```python
#the equation is 
#A00*F1+A01*F2+A02*F3=B0
#A10*F1+A11*F2+A12*F3=B1
#A20*F1+A21*F2+A22*F3=B2
```

```python
def Get_R(xbj,Q,z):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    R=5*(u+d+ub+db)/(4*(u+ub)+(d+db))
    return R
```

```python
def A_00(xbj,Q,z,r1m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    return 4*r1m*(u+d)-4*Get_R(xbj,Q,z)*u
def A_01(xbj,Q,z,r1m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return r1m*(ub+db)-Get_R(xbj,Q,z)*db
def A_02(xbj,Q,z,r1m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return 4*r1m*(ub+db)-4*Get_R(xbj,Q,z)*ub
def B_0(xbj,Q,z,r1m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return Get_R(xbj,Q,z)*d-r1m*(u+d)
    
```

```python
def A_10(xbj,Q,z,r2m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return (4*r2m*(u+d) - 4*Get_R(xbj,Q,z)*ub)
def A_11(xbj,Q,z,r2m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return r2m*(ub+db)-Get_R(xbj,Q,z)*d
def A_12(xbj,Q,z,r2m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return 4*r2m*(ub+db)-4*Get_R(xbj,Q,z)*u
def B_1(xbj,Q,z,r2m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return Get_R(xbj,Q,z)*db-r2m*(u+d)
    
```

```python
def A_20(xbj,Q,z,r3m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return (4*r3m*(u+d) - 4*(ub+db))
def A_21(xbj,Q,z,r3m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return r3m*(ub+db)-(u+d)
def A_22(xbj,Q,z,r3m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return 4*r3m*(ub+db)-4*(u+d)
def B_2(xbj,Q,z,r3m):
    u=PDF.xfxQ(2,xbj,Q)
    d=PDF.xfxQ(1,xbj,Q)
    ub=PDF.xfxQ(-2,xbj,Q)
    db=PDF.xfxQ(-1,xbj,Q)
    return (ub+db)-r3m*(u+d)
    
```

```python
for i,row in Hem_file.iterrows():    
    xbj=row['#x']
    z=row['z']
    Q=row['Q']
    r1m=row['h2p_div_d2p_measured']
    r2m=row['h2n_div_d2p_measured']
    r3m=row['d2n_div_d2p_measured']
    A = np.array([[A_00(xbj,Q,z,r1m),A_01(xbj,Q,z,r1m),A_02(xbj,Q,z,r1m)],[A_10(xbj,Q,z,r2m),A_11(xbj,Q,z,r2m),A_12(xbj,Q,z,r2m)],[A_20(xbj,Q,z,r3m),A_21(xbj,Q,z,r3m),A_22(xbj,Q,z,r3m)]])
    B = np.array([B_0(xbj,Q,z,r1m),B_1(xbj,Q,z,r2m),B_2(xbj,Q,z,r3m)])
    F = np.linalg.solve(A,B)
    print(F)
```
